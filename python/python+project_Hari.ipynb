### 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on..... (5 marks)

```python
df['published_month']=df['published_at'].dt.month
months={1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'june',7:'jul',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'}
```

```python
df['published_month']=df['published_at'].dt.month.map(months)
df
```

### 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count. (5 marks)

```python
video_count_by_month=df.groupby('published_month').size()
sorted_video_count=video_count_by_month.sort_values(ascending=False)
print(sorted_video_count)
```

### 9. Find the count of unique video_id, channel_id and channel_title. (3 marks)

```python
unique_video_id_count=df['video_id'].nunique()
unique_channel_id_count=df['channel_id'].nunique()
unique_channel_title_count=df['channel_title'].nunique()
print("Unique Video Ids: ",unique_video_id_count)
print("Unique channel Ids",unique_channel_id_count)
print("Unique channel Titles",unique_channel_title_count)
```

### 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos. (3 marks)

```python
channel_video_counts = df['video_id'].groupby(df['channel_title']).count()
top_10_channels = channel_video_counts.sort_values(ascending=False).head(10)
bottom_10_channels = channel_video_counts.sort_values(ascending=True).head(10)

print("Top 10 Channels with Highest Videos:")
print(top_10_channels)

print("\nBottom 10 Channels with Lowest Videos:")
print(bottom_10_channels)
```

### 11. Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences. (3 marks)

```python
max_likes_title = df.loc[df['likes'].idxmax()]['title']
min_likes_title = df.loc[df['likes'].idxmin()]['title']

print("Video with Maximum Likes:", max_likes_title)
print("Video with Minimum Likes:", min_likes_title)
```

### 12. Find the title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes and write your inferences. (3 marks)

```python
max_dislikes_title = df.loc[df['dislikes'].idxmax()]['title']
min_dislikes_title = df.loc[df['dislikes'].idxmin()]['title']

print("Video with Maximum Dislikes:", max_dislikes_title)
print("Video with Minimum Dislikes:", min_dislikes_title)
```

### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot. (3 marks)

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation = df['view_count'].corr(df['dislikes'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='view_count', y='dislikes', data=df)
plt.title(f'Correlation: {correlation:.2f}')
plt.xlabel('Number of Views')
plt.ylabel('Number of Dislikes')
plt.show()
```

```python
print("Correlation between views and dislikes:", correlation)
```
